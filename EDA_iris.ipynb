```python
import seaborn as sns
```

```python
sns.displot(iris_data,x='SepalWidthCm',hue='Species',multiple='stack')
```

```python
sns.displot(data=iris_data,x='SepalWidthCm',hue='Species',col="Species")
```

```python
sns.jointplot(data=iris_data,x='SepalLengthCm',y='PetalWidthCm',hue='Species')
```

```python
sns.pairplot(data=iris_data,hue='Species')
```

## K - vecinos más cercanos (KNN)
En los gráficos previos se observa que hay algunos cúmulos que agrupa a especies del mismo tipo, esto nos hace pensar que un nuevo dato introducido podría ser vinculado a uno de los cúmulos presentes.

El algoritmo KNN es un algoritmo de clasificación que permite asociar a un dato nuevo a uno de los cúmulos presentes.

**Procedimiento:**
1. Defina el nuevo dato a clasificar y defina el valor de k (debe ser un número impar)
2. Debe identificar a los k puntos más cercanos al nuevo dato.
3. Entre los k puntos más cercanos se hace una votación para saber a que cúmulo pertenece.
4. El nuevo dato queda clasificado por el cúmulo ganador del paso previo.

El algoritmo es sencillo de aplicar, sus desventajas se hacen notar cuando se trabaja con muchos cúmulos o más atributos. Además es necesario tener el cuenta la métrica a considerar. Esto no quita el hecho de que sea una primera alternativa para solucionar problemas de clustering.
