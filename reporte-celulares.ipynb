Creamos el dataframe

```python
sample_data = celulares[['Dia', 'Ingreso']]
sample_data.columns = ['Dia(X)', 'Ingreso(Y)']

df_2 = pd.DataFrame(sample_data)
```

```python
df_2.info()
```

```python
X= df_2[['Dia(X)']]
Y= df_2['Ingreso(Y)']
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X, Y)
```

```python
Y_prediccion = modelo.predict(X)
```

```python
print(Y_prediccion)
```

```python
#Coeficientes:
print("Intercepto B0:",modelo.intercept_.round(5))
print("Pendiente B1:",modelo.coef_[0].round(5))
```

```python
#Metricas:
print("MSE, Error cuadratico medio",mean_squared_error(Y,Y_prediccion))
print("Coeficiente de determinación R2:",r2_score(Y,Y_prediccion))
```

```python
df_2["Proyeccion (Y')"] = Y_prediccion
```

```python
df_2.head(20)
```

```python
#Grafica

plt.figure(figsize=(15,8)) #Donde 8 es el ancho y 5 alto

plt.scatter(X,Y,color='blue', label='Datos Reales')
plt.plot(X,Y_prediccion, color='red', label ='Linea de Regresion')

plt.title("Regresion Lineal")
plt.xlabel("Dia")
plt.ylabel("Ingreso")

plt.legend()
plt.show()
```

```python
#Metricas:
print("MSE, Error cuadratico medio",mean_squared_error(Y,Y_prediccion))
print("Coeficiente de determinación R2:",r2_score(Y,Y_prediccion))
```
